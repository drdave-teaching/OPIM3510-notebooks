```python
df_p = pd.read_csv('/content/poverty-status-by-town-2018.csv')
df_p.head()
```

```python
df_p.info()
```

```python
df_p['Race/Ethnicity'].value_counts()
```

```python
df_p = df_p[df_p['Race/Ethnicity']=='All']
df_p.shape
```

```python
df_p['Poverty Status'].value_counts()
```

```python
df_p.head()
```

```python
df_p = df_p[df_p['Variable']=='Poverty Status']
df_p = df_p[df_p['Measure Type']=='Percent']
df_p.head()
```

```python
df_p_reshaped = df_p.pivot_table(values='Value',
                                 index=['Town', 'Year'],
                                 columns=['Age', 'Poverty Status'])
df_p_reshaped.head()
```

```python
df_p_reshaped.reset_index(inplace=True)
df_p_reshaped.head()
```

```python
df_p_reshaped.columns
```

```python
# prompt: Using dataframe df_h_reshaped: flatten the multi-index for the columns into one string

df_h_reshaped.columns = ["_".join(x) for x in df_h_reshaped.columns.ravel()]
df_p_reshaped.columns = ["_".join(x) for x in df_p_reshaped.columns.ravel()]
```

```python
df_h_reshaped.columns
```

```python
df_h_reshaped.shape
```

```python
df_h_reshaped.info()
```

```python
df_p_reshaped.columns
```

```python
df_p_reshaped.shape
```

```python
import pandas as pd
new_df = pd.merge(left=df_h_reshaped,
                  right=df_p_reshaped,
                  left_on=['Town_', "Year_"],
                  right_on=['Town_', "Year_"],
                            how='left')
print(new_df.shape)
```
