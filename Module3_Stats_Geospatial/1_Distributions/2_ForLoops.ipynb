# Selecting elements from a list manually

```python
myList = ['hello', 'good morning', 'how are you']
i = 2
myList[i]
```

```python
# print the shape (just three elements)
np.shape(myList)[0]
```

# Selecting elements from a list... with a for loop! With slicing.

```python
np.arange(0,5,1) # low, high (not included, step size)
```

```python
for i in np.arange(0, np.shape(myList)[0],1):
  print(myList[i])

print('------------------\n')
print('All done!')
```

# Selecting the elements from a list... in order... no slicing

```python
myList = ['hello', 'good morning', 'how are you']
```

```python
for i in myList:
  print(i)
```

# `np.arange(start, end, step size)`
Remember, `end` is not included!

```python
# this prints a 1D array (similar to list) of numbers
np.arange(0, 7 ,1)
```

```python
myList2 = np.arange(0, 7 ,1)

for i in myList2:
  print(i)
print('\nAll done!')
```

And another way...

```python
for i in np.arange(0, 7 ,1):
  print(i)
print('All done!')
```

# Initializing a variable and reusing it in the loop

```python
current_val = 100 # imagine you have 100 bucks
for i in np.arange(1,6,1): # invested for 5 years
  print('Year:', i)
  current_val = 1.05*(current_val) # and you have it in a savings bond growing 5%
  print('Value:', current_val)
print('This is how much money I have:', current_val)
```

See the power of compound interest? Set it and forget it!

# Nested for loop - advanced!

```python
a = [1,2,3]
b = [4,5,6]
```

```python
# great exam question ... what is c?
c = 0
for a in np.arange(1,3,1): # outer loop (1,2)
  print('this is a:', a)
  print('------------------')
  for b in np.arange(3,5,1): #inner loop (3,4)
    print('this is b:', b)
    c = a + b + c
    print("---")
    print('this is c:', c)
  print('====')
print('all done! c is:', c)
```
