# Vector vs. Raster data
***Vector*** data consists of **points, lines and polygons**. ***Raster data*** consists of **images** - which can be single band (elevation) or multiband (RGB). Each cell in a raster corresponds to a numeric value or category (like average elevation per 2 km grid cell).

<center>

![vector and raster data](https://i.pinimg.com/originals/0f/f0/ea/0ff0eae9df424ecb95b49e09310a16f4.png)

</center>

In the image above - customers are points, streets are lines, parcels are polygons, and elevation/land use are rasters (images). Together, they can help describe the real world. 

Spatial data is not just for geographers or environmentalists - it's also a critical part of business. Knowing where people are and how they interact with the systems/environment around them is vital decision-making. Knowing how to process, create, customize, enrich and blend datasets together it was makes you a valuable biz analyst - how you combine the pieces in a different way to derive value is what will set you apart. 

If you know where customers are relative to points of interest, infrastructure, stores... if you know the attributes of your customers at the census-tract, zip code or 1 km grid cell level (yes, this is possible!) - you win! 

Whenever possible, go out and make your own data instead of paying someone to do the work for you. You'll make your boss happy - use the extra money you saved on data subscriptions to give yourself a raise! ;)

## Spatial Join ("from, to")

### Point to polygon
The point gets all of the attribues of the polygon (for example, this point now knows what country or town it is in!)

![spatial join](https://gisgeography.com/wp-content/uploads/2019/04/Spatial-Join-1265x723.png)

### Intersect
Two features touch at any location

![intersect](https://gisgeography.com/wp-content/uploads/2019/04/Spatial-Join-Intersect-300x183.png)

### Within a distance
Two features are within a set distance

![within a distance](https://gisgeography.com/wp-content/uploads/2019/04/Spatial-Join-Within-Distance-300x183.png)

###Completely within
The join feature is within the target feature

![completely within](https://gisgeography.com/wp-content/uploads/2019/04/Spatial-Join-Completely-Within-300x183.png)

### Identical
Both features match identically

![identical](https://gisgeography.com/wp-content/uploads/2019/04/Spatial-Join-Identical-300x183.png)

### Closest
The join feature is closest to the target feature

![closest/nearest](https://gisgeography.com/wp-content/uploads/2019/04/Spatial-Join-Closest-300x183.png)

## Multiband rasters (with RGB, infrared bands, others)

Everybody is familiar with RGB - it's how we define color! There are sensors that just measure the colors red, green and blue. Values range from 0-255. You can overlay them to make a color composite image.

<center>

![RGB](https://miro.medium.com/max/833/1*Io0pyC1M7aaUJOZSjEFtXA.png)

</center>

These bands come from the [Landsat 8 satellite](https://openweather.co.uk/blog/post/satellite-imagery-landsat-8-and-its-band-combinations) - imagery is often available FOR FREE. You just need to know where to look and how to process it. Here is an example of RGB all together.

Bands | Wavelength (micrometres) | Resolution (metres)
---|---|---
Band 1 - Coastal aerosol| 0.43 – 0.45 |30
Band 2 – Blue | 0.45 – 0.51|30
Band 3 – Green| 0.53 – 0.59|30
Band 4 – Red| 0.64 – 0.67|30
Band 5 – Near Infrared (NIR)|0.85 – 0.88|30
Band 6 – SWIR 1|1.57 – 1.65|30
Band 7 – SWIR 2|2.11 – 2.29|30
Band 8 – Panchromatic|0.50 – 0.68|15
Band 9 – Cirrus|1.36 – 1.38|30
Band 10 – Thermal Infrared (TIRS) 1|10.60 – 11.19|100 * (30)
Band 11 – Thermal Infrared (TIRS) 2|11.50 – 12.51|100 * (30)

Different band locations can make it easier (for the eye) to identify different phenomenon (like fire, smog, vegetation, and others). For example, here it is for identifying crop health (red is dead crops, green are lush crops, yellow are in the middle - think about how this ties to precision agriculture):

<center>
$\text{NDVI} = \text{(NIR - R) / (NIR + R)}$
</center>

NDVI is just an example of a clever way that you can combine band information to derive an insight!

# What's on Tap?
* **Geocoding** - returning the latitude/longitude coordinates based on an address
* **Spatial join** - join latitude/longitude points to a shapefile of zipcodes (or census tracts)
* **Nearest neighbor (and merge?)** - what are the nearest airports to each latitude/longitude point?
* **Merge** - enrich the latitude/longitude data with pre-processed .csv files full of Census data from Github 
* **Model** - build a model based off of the features we created
* **Special treat** - raster data processing for the 2021 Winter Storm that affected Texas. Measure the max(continuous hours below freezing) to identify which areas were hardest hit during the polar vortex.
